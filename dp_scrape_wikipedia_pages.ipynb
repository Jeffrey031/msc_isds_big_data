```python
import pandas as pd
import os
import wikipedia
import time
import datetime
```

```python
### Sample random Wikipedia pages
```

```python
def random_wiki_page():
   random = wikipedia.random(1)
   try:
        wiki_obj = wikipedia.page(random)
   except wikipedia.exceptions.DisambiguationError as e:
        wiki_obj = random_page()
   return wiki_obj
```

```python
def scrape_wiki_obj(obj, dest_folder):
    time.sleep(2) # being nice :D
    if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)
    wiki = obj
    title, text = wiki.title, wiki.content
    title = title.replace(' ', '_')
    if not os.path.exists(os.path.join(dest_folder, title)):
        print(datetime.datetime.now())
        with open(os.path.join(dest_folder, title), "w") as text_file:
            text_file.write(text)
    else:
        print('already scraped {0}'.format(title))
        scrape_wiki_obj(obj, dest_folder)
```

```python
for i in range(1000):
    try:
        scrape_wiki_obj(random_wiki_page(), 'data')
    except:
        print('pass')
        pass
```
